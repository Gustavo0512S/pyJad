```python
# Condição simples - status por avaliação

avaliação = 6

if avaliação >= 7.0: # se verdadeiro or true
    print("boa meu patrão")

else: # se não
    print("Burrão")
```

```python
# Condição composta


if avaliação >= 7.0: # se verdadeiro or true
    print("boa meu patrão")

elif avaliação >= 5.0: # pra usar varias condições (SES)
    print("tamo quase")

else: # se não
    print("Burrão")
```

```python
# Condição Aninhada

if avaliação >= 7.0: # se verdadeiro or true
    print("boa meu patrão")

elif avaliação >= 5.0: # pra usar varias condições (SES)
    resposta = input ("Quer justificar sua resposta? [S or N]: ").upper() # importante definir as condições aceitadas antes de começar o teste
    if resposta == "S":
        print("justifique sua resposta")
    elif resposta == "N":
        print("Ok, zé mané")
    else:
        print("resposta inválida")

else: # se não
    print("Burrão")
```

```python
# Match/Case

menu = int(input("1-Cadastrar | 2-vizualizar | Opção: "))

match menu: # lidar com a resposta direta 
    case 1:
        print("escolheu 1")
    case 2:
        print("escolheu 2")
    case _: # se não
        print("resposta inválida")
```

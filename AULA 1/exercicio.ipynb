```python
nome = input("Nome: ")
cargo = int(input("2. Analista | 1. Assistente | 3. Coordenador | Opção: "))
experiencia = int(input("Anos de experiência: "))
```

```python
salario_assistente = 3550.98 
salario_analista = 5001.25
salario_coordenador = 15890.43
```

```python
match cargo:
    case 1:
        Cargo = "Assistente"
        Salário = salario_assistente
    case 2:
        Cargo = "Analista"
        Salário = salario_analista     
    case 3:
        Cargo = "Coordenador"
        Salário = salario_coordenador
    case _:
        Cargo = "Opção Inválida"
        Salário = 0


```

```python
# REGRAS DE CLASSIFICAÇÃO

if experiencia <= 2 and experiencia >= 0:
    senioridade = 'junior'

elif experiencia <=5 and experiencia >=3:
    senioridade = 'pleno'

elif experiencia >= 6 and experiencia <=40:
     senioridade = "Senior"
   
else:
    senioridade = "erro"
```

```python
# LOGICAS DE DECISÃO

if cargo == 'Analista' and senioridade != 'junior':
    permissao = 'Gerenciamento liberado'

elif cargo == "Assistente" and senioridade == "junior":
    permissao = 'Apenas opções basicas'

elif cargo == "Coordenador" and senioridade == "Senior":
    permissao = "Acesso total"

elif senioridade == 'erro':
    permissao = "Verifique sua resposta"

else: permissao = "Permissão negada"

 





```

```python
print("Nome:", nome)
print("Cargo:", Cargo, senioridade)
print("Acesso usuário;", permissao)
print("Salário:", Salário)
print("Anos de Experiência:", experiencia, "anos")
b
```

```python
senha_cadastrada = input('Senha: ')
senha_entrada = ''
tentativas = 1

while senha_cadastrada != senha_entrada and tentativas < 4:
    senha_entrada = input('Digite sua senha: ')
    tentativas += 1

    if tentativas == 3 and senha_cadastrada != senha_entrada:
        print("Máximo de tentativas atigindas")
   
if senha_cadastrada == senha_entrada:
    print('Acesso permitido')   
   
  

```

```python
while True:
    menu = int(input('1.Cadastrar | 2.Visualizar | 3.Sair | Opção: '))

    match menu:
        case 1:
            nome_franquia = input('Franquia: ')

        case 2: 
            print(nome_franquia)
        
        case 3: 
            break

        case _:
            print('opção invalida')
```

```python
# Entrada de valores

franquias = ['SP 01', 'SP 02', 'SP 03']

pg_franquias = []

for f in franquias: 
    valor = float(input(f'{f} | Valor: '))

    pg_franquias.append(f)
    pg_franquias.append(valor)

pg_franquias
```

```python
# Cálculo

soma = 0 
for indice, dado in enumerate(pg_franquias):
    if indice % 2 == 1:
        soma += dado * 0.01
        
        print(soma)
        
```

```python
franquias = [
    ['SP 01': 50000],
    ['SP 02': 89000],
    ['SP 03': 12350] ]

taxa = 0.10 

print(f"{'SP 01':<20} | {'Faturamento':<12} | {'Percentual(10%)'}")
print("-" * 50)

total_receber = 0

for f in franquias:
    valor_pago = f["faturamento"] * taxa
    total_receber += valor_pago
    print(f"{f['SP 01']:<20} | R$ {f['faturamento']:>10.2f} | R$ {valor_pago:>10.2f}")

print("-" * 50)
print(f"Total a receber no mês: R$ {total_receber:.2f}")
```

```python
franquias_cadastradas = []

for contador in range(2):
    franquia = dict(
        id = int(input('Identificação: ')),
        cpf = input('CPF: '),
        endereco = input('Endereço: ').title(),
        bairro = input("Bairro: ").title(),
        cidade = input("Cidade: ").title(),
        estado = input('Estado: ').upper()
    )
    
    franquias_cadastradas.append(franquia)


for i, franquia in enumerate(franquias_cadastradas, start = 1):
    print(f"\n{i}º Franquia")
    print(f"ID = {franquia['id']}")
    print(f"CPF = {franquia['cpf']}")
    print(f"ENDEREÇO = {franquia['endereco'].upper()}")
    print(f"BAIRRO = {franquia['bairro'].upper()}")
    print(f"CIDADE = {franquia['cidade'].upper()}")
    print(f"ESTADO = {franquia['estado']}")
```

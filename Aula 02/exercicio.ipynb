```python
transacoes = []

while True:
    opcao = input("\nMovimento banc√°rio: 1. Entrada 2. Saida 3. Vizualizar 4. Sair: ")

    if opcao == '1':
        valor = float(input("Valor da entrada: "))
        transacoes.append(f"Entrada: +R$ {valor:.2f}")
        print("Entrada registrada")

    elif opcao == '2':
        valor = float(input("Valor da sa√≠da: "))
        transacoes.append(f"Sa√≠da: -R$ {valor:.2f}")
        print("Sa√≠da registrada")

    elif opcao == '3':
        print("\n--- Hist√≥rico de Transa√ß√µes ---")
        for t in transacoes:
            print(t)
        if not transacoes:
            print("Nenhuma transa√ß√£o registrada.")
            
 
        sub_opcao = input("\n1. Menu inicial, 2. Sair: ")
        if sub_opcao == '2':
            print("Saindo...")
            break

    elif opcao == '4':
        print("Encerrando sistema...")
        break

    else:
        print("Op√ß√£o inv√°lida!")
```

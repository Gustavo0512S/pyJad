Importando bibliotecas


```python
import json
from datetime import datetime
```

Carregando Dados

```python

caminho = (r'dados\contas.json')

try:
    with open(caminho, 'r', encoding = 'UTF8') as arquivo:
        contas = json.load(arquivo)                          

except FileNotFoundError as e:
    print(e)
```

Carregar Contas

```python
def carregar_contas(arquivo):
    
    with open(arquivo, 'r', encoding='utf-8') as f:
        return json.load(f)
```

Converter Datas

```python
def converter_datas(contas):
    for conta in contas:
        conta['vencimento'] = datetime.strptime(conta['vencimento'], '%d/%m/%Y')
```

```python
for conta in contas:                                                                
    conta['vencimento'] = datetime.strptime(conta['vencimento'], "%d/%m/%Y")
 
contas_vencidas = []
contas_vencidas_pagar =[]
contas_vencidas_receber = []
contas_erro_tipo = []
 
for conta in contas:
    if conta['vencimento'] < datetime.today():
        contas_vencidas.append(conta)
 
        if conta['tipo'] == "Pagar":
            contas_vencidas_pagar.append(conta)
 
        elif conta['tipo'] == "Receber":
            contas_vencidas_receber.append(conta)
 
        else:
            contas_erro_tipo.append(conta)
 
print(len(contas_vencidas))
print(len(contas_vencidas_pagar))
print(len(contas_vencidas_receber))
print(len(contas_erro_tipo))
 
```

Identificar contas vencidas

```python

def identificar_contas_vencidas(contas):

    hoje = datetime.today()
    contas_vencidas = []
    
    for conta in contas:
        if conta['vencimento'] < hoje and conta['status'] == 'Em aberto':
            contas_vencidas.append(conta)
    
    return contas_vencidas
```

Total a pagar

```python

def calcular_total_pagar(contas):

    total = 0
    for conta in contas:
        if conta['tipo'] == 'Pagar' and conta['status'] == 'Em aberto':
            total += conta['valor']
    return total
```

Total a receber

```python
def calcular_total_receber(contas):
    
    total = 0
    for conta in contas:
        if conta['tipo'] == 'Receber' and conta['status'] == 'Em aberto':
            total += conta['valor']
    return total
```

Imprimir Resumo

```python
def exibir_resumo(contas, contas_vencidas, total_pagar, total_receber):
  
    saldo = total_receber - total_pagar
    
    print("=" * 60)
    print("RESUMO FINANCEIRO".center(60))
    print("=" * 60)
    print(f"\n Total a Pagar: R$ {total_pagar:,.2f}")
    print(f"Total a Receber: R$ {total_receber:,.2f}")
    print(f" Contas Vencidas: {len(contas_vencidas)}")
    print(f"\n{'Saldo Positivo' if saldo >= 0 else 'Saldo Negativo'}: R$ {saldo:,.2f}")
    print("=" * 60)
    
    if contas_vencidas:
        print("\n CONTAS VENCIDAS:")
        print("-" * 60)
        for conta in contas_vencidas:
            print(f"ID: {conta['id']} | {conta['descricao']} | R$ {conta['valor']:,.2f} | Venc: {conta['vencimento'].strftime('%d/%m/%Y')}")
        print("-" * 60)
```

```python
# 1. Carregar dados do arquivo JSON
contas = carregar_contas('contas.json')

# 2. Converter datas de string para datetime
converter_datas(contas)

# 3. Identificar contas vencidas
contas_vencidas = identificar_contas_vencidas(contas)

# 4. Calcular totais financeiros
total_pagar = calcular_total_pagar(contas)
total_receber = calcular_total_receber(contas)

# 5. Exibir resumo consolidado
exibir_resumo(contas, contas_vencidas, total_pagar, total_receber)
```

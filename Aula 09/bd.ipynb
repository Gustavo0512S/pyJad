Nome dos clientes

```python
query_cols_clientes = """
SELECT COLUMN_NAME
FROM INFORMATION_SCHEMA.COLUMNS
WHERE TABLE_NAME = 'clientes'
"""

pd.read_sql(query_cols_clientes, conn)

query_clientes = """
SELECT
    nomeCliente
FROM dbo.clientes
ORDER BY nomeCliente
"""

df_clientes = pd.read_sql(query_clientes, conn)
df_clientes


```

Produtos com valor > 2000


```python
query_cols_produtos = """
SELECT COLUMN_NAME
FROM INFORMATION_SCHEMA.COLUMNS
WHERE TABLE_NAME = 'produtos'
"""

pd.read_sql(query_cols_produtos, conn)
```

```python
query_produtos = """
SELECT
    nomeProduto,
    descricaoProduto,
    precoUnitario
FROM dbo.produtos
WHERE precoUnitario > 2000
ORDER BY precoUnitario DESC
"""

df_produtos = pd.read_sql(query_produtos, conn)
df_produtos
```
